```python
cos = torch.nn.CosineSimilarity(dim=0, eps=1e-6)
```

```python
def find_k_nearest_neighbors(word, embedding_dict, k):
  words_cosine_similarity = dict()
  for token in embedding_dict.keys():
    words_cosine_similarity[token] = cos(embedding_dict[word], embedding_dict[token]).item()
  words_cosine_similarity = dict(sorted(words_cosine_similarity.items(), key=lambda item: item[1]))
  return list(words_cosine_similarity.keys())[-k:][::-1]

def delete_hashtag_usernames(text):
  try:
    result = []
    for word in text.split():
      if word[0] not in ['@', '#']:
        result.append(word)
    return ' '.join(result)
  except:
    return ''

def delete_url(text):
  text = re.sub(r'http\S+', '', text)
  return text
```

```python
# 1. find one hot encoding of each word for each year
# 2. find 10 nearest words from "ولنتاین"
word2ind = {w:i for i,w in enumerate(final_unique)}
```

#2. TF-IDF

```python
# 1. find the TF-IDF of all tweets.
# 2. choose one tweets randomly.
# 3. find 10 nearest tweets from chosen tweet.

sentences = tweets_final
# for t in tweets_final:
#   sentences.append(t.split(" "))

doc_num = len(sentences)

word_sen_counts = [0]*word_num
word_counts = [0]*word_num

tfs = []
tf_idf = {}

for sen2 in sentences:
  my_sen = sen2.split(" ")
  l_sen = len(my_sen)
  visited = []
  for word in my_sen:
    word_counts[word2ind[word]] += 1
    if word not in visited:
      visited.append(word)
      word_sen_counts[word2ind[word]] += 1
  for i in range(len(word_counts)):
    word_counts[i] /= l_sen
  tfs.append(word_counts)

for i,sen in enumerate(sentences):
  my_sen = sen.split(" ")
  val = [0]*word_num
  for j,word in enumerate(my_sen):
    val[word2ind[word]] = tfs[i][word2ind[word]] * np.log(doc_num/word_sen_counts[word2ind[word]])


  tf_idf[sen] = torch.from_numpy(np.array(val))
```

```python
find_k_nearest_neighbors('زن، زندگی ،ازادی', tf_idf, 10)
```

```python
from transformers import BertModel, BertTokenizer
```

```python
# 1. fine tune a bert model base on all tweets for each year.
# 2. find 10 nearest words from "ولنتاین"

model = BertModel.from_pretrained(model_checkpoint, output_hidden_states = True)
tokenizer = BertTokenizer.from_pretrained(model_checkpoint)
```

```python
def bert_text_preparation(text, tokenizer):
  marked_text = "[CLS] " + text + " [SEP]"
  tokenized_text = tokenizer.tokenize(marked_text)
  indexed_tokens = tokenizer.convert_tokens_to_ids(tokenized_text)
  segments_ids = [1]*len(indexed_tokens)
  tokens_tensor = torch.tensor([indexed_tokens])
  segments_tensor = torch.tensor([segments_ids])
  return tokenized_text, tokens_tensor, segments_tensor
```

```python
def get_bert_embeddings(tokens_tensor, segments_tensor, model):
  with torch.no_grad():
    outputs = model(tokens_tensor, segments_tensor)
    hidden_states = outputs[2]
  token_embeddings = torch.stack(hidden_states, dim=0)
  token_embeddings = torch.squeeze(token_embeddings, dim=1)
  token_embeddings = token_embeddings.permute(1,0,2)
  token_vecs_sum = []

  for token in token_embeddings:
    token_vecs_sum.append(token[-1])
  return token_vecs_sum
```

```python
bert_embed = {}
count = 0

for sen in tweets_final:
  count += 1
  if count % 100 == 0:
    print(f"{count} / {len(tweets_final)}")
  # print(sen)
  tokenized_text, tokens_tensor, segments_tensor = bert_text_preparation(sen, tokenizer)
  # print(len(tokenized_text))
  token_vec = get_bert_embeddings(tokens_tensor, segments_tensor, model)
  # print(tokenized_text)
  for i,token in enumerate(tokenized_text[1:-1]):
    bert_embed[token] = token_vec[i]
```

```python
find_k_nearest_neighbors('زندگی', bert_embed, 10)
```
